```
#For Steel
import matplotlib.pyplot as plt
import csv
import math
import pandas as pd
import numpy as np

Force = []
Extension = []

with open('Steel.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        Force.append(float(row[0]))
        Extension.append(float(row[1]))
Stress = []
for Forces in Force:
  Stress.append((Forces*4)/(math.pi*0.007*0.007))

Strain = []
for extensions in Extension:
  Strain.append(extensions/100)

True_Stress = []
True_Strain = []

for i in range (0, 980):
  x = Stress[i] * (1 + Strain[i])
  i = i + 1
  True_Stress.append(x)

for i in range (0, 980):
  y = math.log(1+Strain[i])
  i = i + 1
  True_Strain.append(y)

j = Stress[400]
k = Strain[400]
h = Stress[800]
i = Strain[800]
Slope = (h-j)/(i-k)
print(Slope)

plt.plot(Strain, Stress)
plt.plot(True_Strain, True_Stress)
plt.scatter(k, j)
plt.scatter(i, h)
plt.title('Steel')
plt.xlabel('Strain')
plt.ylabel('Stress in N/m2')
plt.show()

plt.plot(Extension, Force)
plt.title('Force vs Displacement Curve')
plt.xlabel('Displacement in mm')
plt.ylabel('Force in Newton')
a = Force[400]
b = Extension[400]
print(b, a)
c = Force[800]
d = Extension[800]
print(d, c)
Slope = (c-a)/(d-b)
print(Slope)
plt.scatter(b, a)
plt.scatter(d, c)
plt.show()
```
